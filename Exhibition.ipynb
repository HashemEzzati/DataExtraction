```python
import pandas as pd
import Funpat as fp
import numpy as np
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc

ExtractedData = pd.read_csv('ExtractedData.csv')
    
########**  Counting Labels of Extracted Data  **########            
CouLab = ExtractedData.groupby(by=['Label']).size().reset_index(name='Count').sort_values(['Count'], ascending=True,ignore_index=True)
            
                
########**  Data Exhibition **########

rows = len(CouLab)
cols = 1
fig, ax = plt.subplots(rows, cols, figsize=(10, 90))


for i in range(0,rows):
    rowindex = ExtractedData[ExtractedData.Label==CouLab['Label'][i]].first_valid_index()
    for j in range(0,cols):
        x = np.append(np.array([range(0,10)]).reshape(10,1),ExtractedData.loc[rowindex+j][:].drop('Label').to_numpy().reshape(10,4),axis=1)
        candlestick_ohlc(ax[i],x,width=0.6, colorup = 'green', colordown = 'red', alpha = 0.8)
        ax[i].set_xlabel(CouLab['Label'][i])
plt.show() 
```
